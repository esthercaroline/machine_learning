```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([ 
    [1],
    [5],
    [2],
    [4],
])

y = np.array([
    [1],
    [2],
    [1.3],
    [1.7],
])
```

```python
plt.figure()
plt.scatter(X, y)
plt.show()
```

```python
x_plot = np.linspace(0, 6, 100).reshape(-1, 1)

theta_0 = 0.75
theta_1 = 0.25

y_plot = theta_0 + x_plot * theta_1
```

```python
plt.figure()
plt.scatter(X, y)
plt.plot(x_plot, y_plot, color='red')
plt.show()
```

```python
A = np.array([
    [2, 6],
    [24, 92],
])

b = np.array([
    [3],
    [40.8],
])
```

```python
np.linalg.solve(A, b)
```

```python
np.linalg.inv(A) @ b
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

model.intercept_, model.coef_
```

```python
x_plot = np.linspace(0, 6, 100).reshape(-1, 1)
y_plot = model.predict(x_plot)
```

```python
plt.figure()
plt.scatter(X, y)
plt.plot(x_plot, y_plot, color='red')
plt.show()
```
